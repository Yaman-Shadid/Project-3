```python
# Combine both data frames
combined_df = pd.concat([df_2000, df_2001])
combined_df
```

```python
# movies that had at least some valid financial information (values > 0 for budget OR revenue)

valid_financial_movies = combined_df[(combined_df['budget'] > 0) | (combined_df['revenue'] > 0)]
count_valid_financial_movies = len(valid_financial_movies)
print(f"Number of movies with valid financial information: {count_valid_financial_movies}")
```

```python
# Exclude any movies with 0's for budget AND revenue from the remaining visualizations.
filtered_combined_df = combined_df[(combined_df['budget'] > 0) & (combined_df['revenue'] > 0)]
```

```python
# How many movies are there in each of the certification categories (G/PG/PG-13/R)

certification_counts = filtered_combined_df['certification'].value_counts()
print(certification_counts)
```

```python
# What is the average revenue per certification category

average_revenue_per_certification = filtered_combined_df.groupby('certification')['revenue'].mean()
print(average_revenue_per_certification)
```

```python
# What is the average budget per certification category

average_budget_per_certification = filtered_combined_df.groupby('certification')['budget'].mean()
print(average_budget_per_certification)
```
