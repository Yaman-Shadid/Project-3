## Hypothesis Testing 1

- Does the MPAA rating of a movie (G/PG/PG-13/R) affect how much revenue the movie generates?

```python
import scipy.stats as stats

# Filter dataframe for non-zero revenues and where certification is present
filtered_df = df[(df['revenue'] > 0) & (df['certification'].notnull())]

# Create a list of revenue arrays for each movie rating
rating_revenues = [
    filtered_df[filtered_df['certification'] == 'G']['revenue'],
    filtered_df[filtered_df['certification'] == 'PG']['revenue'],
    filtered_df[filtered_df['certification'] == 'PG-13']['revenue'],
    filtered_df[filtered_df['certification'] == 'R']['revenue']
]

# Perform Kruskal-Wallis test
h_stat, p_value = stats.kruskal(*rating_revenues)

print(f"H-statistic: {h_stat:.2f}")
print(f"P-value: {p_value:.5f}")

# Interpret the results
if p_value < 0.05:
    print("There is a statistically significant difference in revenue among movie ratings.")
else:
    print("There is no statistically significant difference in revenue among movie ratings.")
```

```python
# Use Shapiro-Wilk test to check normality for each group
for i, group in enumerate(rating_revenues):
    _, p_value = stats.shapiro(group)
    print(f"Shapiro-Wilk p-value for rating '{rating_revenues[i].name}': {p_value:.5f}")
```

## Hypothesis 2

- Do movies that are over 2.5 hours long earn more revenue than movies that are 1.5 hours long (or less)?

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import scipy.stats as stats

# Filter dataframe for non-zero revenues and where certification is present
filtered_df = df[(df['revenue'] > 0) & (df['certification'].notnull())]

# Separate movies based on runtime
long_movies = filtered_df[filtered_df['runtime'] > 150]['revenue']
short_movies = filtered_df[filtered_df['runtime'] <= 90]['revenue']

# Check the assumption of independence
# Check the assumption of normality
# Use Shapiro-Wilk test to check normality for both groups
_, p_value_long = stats.shapiro(long_movies)
_, p_value_short = stats.shapiro(short_movies)
print(f"Shapiro-Wilk p-value for long movies: {p_value_long:.5f}")
print(f"Shapiro-Wilk p-value for short movies: {p_value_short:.5f}")

# Perform T-test
t_stat, p_val = stats.ttest_ind(long_movies, short_movies, equal_var=False)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.5f}")

# Interpret the results
if p_val < 0.05:
    print("There is a statistically significant difference in revenue between long and short movies.")
else:
    print("There is no statistically significant difference in revenue between long and short movies.")

# Plot histograms for each group to visualize the distribution
plt.figure()
plt.hist(long_movies, bins=20, alpha=0.5, label='Long Movies')
plt.hist(short_movies, bins=20, alpha=0.5, label='Short Movies')
plt.title('Revenue distribution for Long vs. Short Movies')
plt.legend()
plt.show()
```

## Hypothesis 3

- Do movies released in 2020 earn less revenue than movies released in 2018?

```python
import scipy.stats as stats
import matplotlib.pyplot as plt
import scipy.stats as stats

# Filter dataframe for non-zero revenues and where certification is present
filtered_df = df[(df['revenue'] > 0) & (df['certification'].notnull())]

# Extract year from release_date (assuming it's in 'YYYY-MM-DD' format)
filtered_df['release_year'] = pd.to_datetime(filtered_df['release_date']).dt.year

# Separate movies based on release year
movies_2020 = filtered_df[filtered_df['release_year'] == 2020]['revenue']
movies_2018 = filtered_df[filtered_df['release_year'] == 2018]['revenue']

# Check the assumption of independence
# Check the assumption of normality
# Use Shapiro-Wilk test to check normality for both groups
_, p_value_2020 = stats.shapiro(movies_2020)
_, p_value_2018 = stats.shapiro(movies_2018)
print(f"Shapiro-Wilk p-value for movies released in 2020: {p_value_2020:.5f}")
print(f"Shapiro-Wilk p-value for movies released in 2018: {p_value_2018:.5f}")

# Perform T-test
t_stat2, p_val2 = stats.ttest_ind(movies_2020, movies_2018, equal_var=False)
print(f"T-statistic: {t_stat2:.2f}, P-value: {p_val2:.5f}")

# Interpret the results
if p_val2 < 0.05:
    print("There is a statistically significant difference in revenue between movies released in 2020 and 2018.")
else:
    print("There is no statistically significant difference in revenue between movies released in 2020 and 2018.")

# Plot histograms for each group to visualize the distribution
plt.figure()
plt.hist(movies_2020, bins=20, alpha=0.5, label='Movies Released in 2020')
plt.hist(movies_2018, bins=20, alpha=0.5, label='Movies Released in 2018')
plt.title('Revenue distribution for Movies Released in 2020 vs. 2018')
plt.legend()
plt.show()
```
