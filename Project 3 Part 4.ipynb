## Imports

```python
import pandas as pd
```

```python
# Read the dataframes

tmdb_2018 = pd.read_csv(r'Data\tmdb_2018.csv.gz')
tmdb_2019 = pd.read_csv(r'Data\tmdb_2019.csv.gz')
tmdb_2020 = pd.read_csv(r'Data\tmdb_2020.csv.gz')
tmdb_2021 = pd.read_csv(r'Data\tmdb_2021.csv.gz')
```

```python
# Combine the dataframes
df = pd.concat([tmdb_2018, tmdb_2019, tmdb_2020, tmdb_2021], ignore_index=True)
```

```python
df.head()
```

```python
df = df.loc[df['imdb_id']!='0']
df
```

```python
df = df.reset_index(drop=True)
```

```python
df.head()
```

## Hypothesis Testing 1

- Does the MPAA rating of a movie (G/PG/PG-13/R) affect how much revenue the movie generates?

```python
import scipy.stats as stats

# Filter dataframe for non-zero revenues and where certification is present
filtered_df = df[(df['revenue'] > 0) & (df['certification'].notnull())]

# Extract revenues for each movie rating
g_revenue = filtered_df[filtered_df['certification'] == 'G']['revenue']
pg_revenue = filtered_df[filtered_df['certification'] == 'PG']['revenue']
pg13_revenue = filtered_df[filtered_df['certification'] == 'PG-13']['revenue']
r_revenue = filtered_df[filtered_df['certification'] == 'R']['revenue']

# Perform ANOVA test
f_stat, p_value = stats.f_oneway(g_revenue, pg_revenue, pg13_revenue, r_revenue)

print(f"F-statistic: {f_stat:.2f}")
print(f"P-value: {p_value:.5f}")
```

### Visualization

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 6))
sns.boxplot(data=filtered_df, x='certification', y='revenue', order=['G', 'PG', 'PG-13', 'R'])
plt.title("Distribution of Revenue across Movie Ratings")
plt.ylabel("Revenue")
plt.xlabel("Movie Ratings")
plt.show()
```

## Hypothesis 2

- Do movies that are over 2.5 hours long earn more revenue than movies that are 1.5 hours long (or less)?

```python
# Separate movies based on runtime
long_movies = filtered_df[filtered_df['runtime'] > 150]['revenue']
short_movies = filtered_df[filtered_df['runtime'] <= 90]['revenue']

# T-test
t_stat, p_val = stats.ttest_ind(long_movies, short_movies, equal_var=False)
print(f"T-statistic: {t_stat:.2f}, P-value: {p_val:.5f}")
```

### Visualization

```python
# Prepare data
long_movies_df = filtered_df[filtered_df['runtime'] > 150]
short_movies_df = filtered_df[filtered_df['runtime'] <= 90]

runtime_df = pd.concat([long_movies_df.assign(category='> 2.5 hours'), short_movies_df.assign(category='<= 1.5 hours')])

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=runtime_df, x='category', y='revenue')
plt.title("Revenue Distribution: Movies > 2.5 hours vs. <= 1.5 hours")
plt.ylabel("Revenue")
plt.xlabel("Runtime Category")
plt.show()
```

## Hypothesis 3

- Do movies released in 2020 earn less revenue than movies released in 2018?

```python
# Extract year from release_date (assuming it's in 'YYYY-MM-DD' format)
filtered_df['release_year'] = pd.to_datetime(filtered_df['release_date']).dt.year

# Separate movies based on release year
movies_2020 = filtered_df[filtered_df['release_year'] == 2020]['revenue']
movies_2018 = filtered_df[filtered_df['release_year'] == 2018]['revenue']

# T-test
t_stat2, p_val2 = stats.ttest_ind(movies_2020, movies_2018, equal_var=False)
print(f"T-statistic: {t_stat2:.2f}, P-value: {p_val2:.5f}")
```

### Visualization

```python
# Prepare data (assuming you've already created the 'release_year' column)
movies_2020_df = filtered_df[filtered_df['release_year'] == 2020]
movies_2018_df = filtered_df[filtered_df['release_year'] == 2018]

release_df = pd.concat([movies_2020_df.assign(year='2020'), movies_2018_df.assign(year='2018')])

# Boxplot
plt.figure(figsize=(10, 6))
sns.boxplot(data=release_df, x='year', y='revenue')
plt.title("Revenue Distribution: Movies in 2020 vs. 2018")
plt.ylabel("Revenue")
plt.xlabel("Release Year")
plt.show()
```
